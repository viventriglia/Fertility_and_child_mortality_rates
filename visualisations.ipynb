```python
import matplotlib.pyplot as pp
import pandas as pd
import numpy as np

from IPython import display
from ipywidgets import interact, widgets
from ipywidgets.embed import embed_minimal_html

%matplotlib inline
```

```python
gapminder = pd.read_csv('gapminder.csv')
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year]
    area = 5e-6 * data.population
    colours = data.region.map({'Africa':'skyblue','Europe':'gold','America':'palegreen', 'Asia':'coral'})
    
    data.plot.scatter('babies_per_woman', 'age5_surviving',
                     s=area, c=colours, linewidths=1, edgecolors='k', figsize=(12,9))
    
    pp.axis(xmin=0, xmax=8, ymin=50, ymax=105)
    pp.xlabel('babies per woman')
    pp.ylabel('% of children alive at the age of 5')
    pp.title(f'{year}')
```

```python
interact(plotyear, year=widgets.IntSlider(min=1950, max=2015, step=1, value=1970))
```

```python
for year in range(1950,2016,1):
    plotyear(year)
    pp.savefig(f'images/{year}.png', bbox_inches="tight", dpi=72)
```
